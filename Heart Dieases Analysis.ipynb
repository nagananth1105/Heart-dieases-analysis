**AFFECTED AND UNAFFECTED BY HEART DIEASES**

```python
data.columns
```

```python
import matplotlib.pyplot as plt
sns.countplot(x='target',data=data)
plt.title('AFFECTED AND UNAFFECTED BY HEART DIEASES')
plt.xticks([0,1],['unaffected','affected'])
plt.show()
```

**Gender Classification**

```python
data.columns
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x='sex',data=data)
plt.title('Gender Classification')
plt.xticks([0,1],['female','male'])
plt.show()
```

```python
gender_target_counts = data.groupby(['sex', 'target']).size().unstack(fill_value=0)
gender_target_percentage = gender_target_counts.div(gender_target_counts.sum(axis=1), axis=0) * 100
print(gender_target_percentage)
plt.figure(figsize=(8, 6))
gender_target_percentage.plot(kind='bar', stacked=True)
plt.title('Percentage of Heart Disease by Gender')
plt.xlabel('Gender (0: Female, 1: Male)')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.legend(title='Target (Heart Disease)')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.displot(data['age'], color='green')
plt.title("Age classification")
plt.show()
```
